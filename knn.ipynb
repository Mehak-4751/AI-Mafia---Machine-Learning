```python
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(x,y,querypt,k=5):
    vals=[]
    for i in range(x.shape[0]):
        dist=distance(x[i],querypt)
        vals.append((dist,y[i]))
    
    #print(vals)
    vals=sorted(vals)
    vals=vals[:k]
    vals=np.array(vals)
    x=np.unique(vals[:,1],return_counts=True)
    #print(x)
    index=x[1].argmax()
    #print(index)
    return x[0][index]
```

```python
#knn(x,y,[5,5])
knn(x,y,querypoint)
```

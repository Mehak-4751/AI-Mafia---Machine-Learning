```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url="https://en.wikipedia.org/wiki/A._P._J._Abdul_Kalam"
```

```python
data=requests.get(url)
clean_data=BeautifulSoup(data.content,"lxml")
print(clean_data)
```

```python
summary=clean_data.findAll(class_='infobox vcard')
```

```python
summary
```

```python
ele=summary[0].findAll('th')
```

```python
x=[i.get_text() for i in ele]
```
